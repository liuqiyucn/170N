```python
# Write program that finds each even integer greater than 57 and less than 97,
#and each time it gets to one of these numbers, it prints out `This is a fun number!`.

i = 57 #first i value
last = 97

while i<last:
  if i%2 == 0:
    print(i, "This is a fun number!")

  i+=1
```

```python
#Write a program that prints out the first 10 Fibonnacci numbers.


prev1 = 0
prev2 = 1
print(1) #this is our starting number

for i in range(9):
  num = prev1+prev2
  print(num)
  prev1 = prev2
  prev2=num
```

```python
#Write a program in which the user can specify a positive integer n,
#and the program will output the nth Fibonacci number.

def getNthFibonacci(n):
  prev1 = 0
  prev2 = 1
  num = 1
  for i in range(n-1):
    num = prev1+prev2
    prev1 = prev2
    prev2=num
  return num


print(getNthFibonacci(10))
```

### Answers -- writing functions

Create your new cell(s) below this one.

```python
#define a function called my_fifth_power that takes a number
#as input and outputs the fifth power of that number.

def my_fifth_power(num):
  nth_power = num
  for n in range(4):
    nth_power = nth_power*num
  return nth_power

print(my_fifth_power(2))
print(my_fifth_power(3))
```

```python
#Define a function my_power that uses only Python's built-in multiplication
#to compute any non-negative integer power of a number.
#thus function should have two inputs, one for the integer and one for the power.

def my_power(num, power):
  nth_power = num
  for n in range(power-1):
    nth_power = nth_power*num
  return nth_power

print(my_power(2,3))
print(my_power(2,5))
print(my_power(3,5))
```

```python
#Define a function my_absolute that takes a float as its input and outputs its absolute value.

def my_absolute(num):
  if num<0:
    num = num*-1
  return num

print(my_absolute(2))
print(my_absolute(-2))
```

```python
#Define a function my_max that takes two numbers as inputs and outputs the larger number if they are different.
#If the two numbers are the same, the function should output either of the two numbers.

def my_max(num1,num2):
  if num1>num2:
    return num1
  return num2


print(my_max(2,1))
print(my_max(-5,3))
print(my_max(2,2))
```

```python
#Define a function my_max_three that takes three numbers as inputs and outputs their maximum.

def my_max_three(num1,num2,num3):
  return my_max(my_max(num1,num3),num3)

print(my_max_three(1,2,3))
print(my_max_three(-10,-10,2))
print(my_max_three(6,-1,6))
print(my_max_three(0,-1,7))
```

```python
#Define a function my_integer_sum that takes a positive integer n as its input and
#outputs the sum of all positive integers less than or equal to n.

def my_integer_sum(n):
  sum = 0
  for i in range(n+1):
    sum+=i
  return sum

print(my_integer_sum(4))
print(my_integer_sum(5))

```

```python
#Define a function my_integer_sum_power that takes a positive integer n and an optional positive integer p
#whose default value is  1  as its input and outputs the sum of all powers  𝑘^𝑝  where
#𝑘  ranges through the positive integers less than or equal to n.


def my_integer_sum_power(n, p=1):
  sum=0
  for k in range(n+1):
    sum+=my_power(k,p)
  return sum

print(my_integer_sum_power(5))
print(my_integer_sum_power(3,2))
```

```python
# Define a function my_integer_product that takes a positive integer n as its input and
# computes the product of all positive integers less than or equal to n as its output.

def my_integer_product(n):
  cur_product = 1
  cur_int = 1
  while cur_int <= n:
    cur_product *= cur_int
    cur_int+=1
  return cur_product

print(my_integer_product(5))

```

```python
# Define a function my_function_sum that takes a float-valued function  𝑓  of integer argument
#and an integer  𝑛  as its inputs, and outputs the sum  𝑓(1)+𝑓(2)+⋯+𝑓(𝑛)  as its output.

def my_function_sum(func, n):
  i = 1
  sum = 0
  while i<=n:
    sum += func(i)
    i+=1
  return sum

print(my_function_sum(my_absolute,5))
```

```python
#Define a function my_function_product that takes a float-valued function  𝑓  of integer argument
#and an integer  𝑛  as its inputs, and outputs the product  𝑓(1)⋅𝑓(2)⋯𝑓(𝑛)  as its output.

def my_function_product(func, n):
  i = 1
  product = 1
  while i<=n:
    product *= func(i)
    i+=1
  return product

print(my_function_product(my_absolute,5))
```

```python
#Define a function my_function_product_powers that takes a float-valued function  𝑓  of integer argument,
# an integer  𝑛 , and an optional integer  𝑝  whose default value is  1  as its inputs, and outputs the
#product  𝑓(1)𝑝⋅𝑓(2)𝑝⋯𝑓(𝑛)𝑝  as its output.


def my_function_product_powers(func, n, p=1):
  i = 1
  product = 1
  while i<=n:
    product *= my_power(func(i),p)
    i+=1
  return product

print(my_function_product_powers(my_absolute,5))
print(my_function_product_powers(my_absolute,5,p=2))
```
