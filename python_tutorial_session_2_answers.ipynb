### Exercises - systematic list creation and manipulation

```python
def first_100_odd():
    return [x for x in range(1, 100+1) if x%2==1]
print(first_100_odd())
```

```python
def first_100_perfect():
    number = 1
    l = []
    while number**2 <= 100:
        l.append(number**2)
        number = number + 1
    return l

print(first_100_perfect())
```

```python
def first_100_perfect_even():
    perfect = first_100_perfect()
    return [x for x in perfect if x % 2 == 0]

print(first_100_perfect_even())
```

```python
def divisible_357():
    return [x for x in range(1, 101) if x % 3 == 0 and x%5==0 and x%7==0]

print(divisible_357())
```

```python
def fib_list(n):
    l = [1, 1]
    pre_index = 0
    after_index = 1
    for _ in range(n-2):
        l.append(l[pre_index] + l[after_index])
        pre_index = pre_index+1
        after_index = after_index+1
    return l

print(fib_list(10))
```

```python
def fib_custom(n):
    l = fib_list(n)
    return [(x**2+1-1/x) for x in l]

print(fib_custom(3))
```

```python
def position():
    v = 5
    time = 0
    out = []
    while time < 20:
        out.append(v*time)
        time = time+0.3
    return out

print(position())
```

### Exercises - 2D plot creation

```python
import matplotlib.pyplot as plt
angle = np.linspace(-3*np.pi, 5*np.pi, 100)
sin = np.sin(angle)
cos = np.cos(angle)
plt.plot(angle, sin, color = 'red')
plt.plot(angle, cos, linestyle = 'dashed', color = 'green')
plt.xlabel('angle')
plt.ylabel('function value')
plt.show()
```

```python
t = np.linspace(0, 2, 400)
frequencies = range(1, 10)
colors = np.linspace([0, 0, 1], [1, 0, 0], len(frequencies))

for freq, color in zip(frequencies, colors):
    y = np.sin(freq * np.pi * t)
    plt.plot(t, y, color=color, label=f'n={freq}')

plt.legend()
plt.xlabel('angle')
plt.ylabel('function value')
plt.show()
```
